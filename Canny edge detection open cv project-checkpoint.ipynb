### Canny Edge detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def canny_edge_detection(image_path, low_threshold, high_threshold):
    image = cv2.imread(image_path)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 1.4)
    edges = cv2.Canny(blurred_image, low_threshold, high_threshold)
    
    return edges

if __name__ == "__main__":
    img_path = "C:\\Users\\just4\\anaconda3\\IMG-20240107-WA0141.jpg"
    edges = canny_edge_detection(img_path, 50, 150)
    plt.imshow(edges, cmap='gray')
    plt.title('Canny Edge Detection')
    plt.axis('off')
    plt.show()
```

### Enhanced Canny Edge detection 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def enhance_canny_edges(image_path):
    image = cv2.imread(image_path)
    if image is None:
        print("Error: Image not found.")
        return
    
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
    
    edges = cv2.Canny(blurred_image, 100, 200)

    enhanced_edges = cv2.convertScaleAbs(edges, alpha=1.5, beta=0)

    combined_image = cv2.addWeighted(image, 0.5, cv2.cvtColor(enhanced_edges, cv2.COLOR_GRAY2BGR), 1, 0)

    plt.imshow(cv2.cvtColor(combined_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')  
    plt.title('Enhanced Canny Edges')
    plt.show()

if __name__ == "__main__":
    input_image_path = r"C:\Users\just4\anaconda3\IMG-20240107-WA0141.jpg"
    enhance_canny_edges(input_image_path)
```
